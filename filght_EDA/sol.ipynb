```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("flights")
df.sample(5)
```

```python
new=df.groupby(['year', 'passengers']).size().reset_index()
```

```python
sns.lineplot(x=new['year'], y=new['passengers'])
```

```python
pivot_table = df.pivot_table(values='passengers',index='month',columns='year')
print(pivot_table)
```

```python
sns.clustermap(pivot_table, cmap='viridis', figsize=(8, 8))
```
