```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips.sample(5)
```

```python
tips["tip"].duplicated().sum()
```

```python
correlation_matrix = tips["tip"].corr(tips["total_bill"])
print(correlation_matrix)
```

```python
# comparing numerical - numerical data
sns.scatterplot(
    data=tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size"
)
```
