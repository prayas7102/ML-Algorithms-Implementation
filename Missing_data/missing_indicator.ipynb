```python
from sklearn.impute import MissingIndicator
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
import pandas as pd

# Sample data

excel_file_path = "./titanic_toy.csv"
df = pd.read_csv(
    excel_file_path, encoding="latin-1", usecols=["Survived", "Age", "Fare"]
)

# Create a missing indicator
# missing_indicator = MissingIndicator()
# missing_df_indicator = missing_indicator.fit_transform(data)
# print(missing_df_indicator)
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["Survived", "Fare"])
y = df["Survived"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
X_train.head()
```

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf, y_train)

y_pred = clf.predict(X_test_trf)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
mi = MissingIndicator()
mi.fit(X_train)
mi.features_
X_train_missing = mi.transform(X_train)
X_train_missing = [x[0] for x in X_train_missing]
X_train_missing
```

```python
X_test_missing = mi.transform(X_test)
X_test_missing = [x[0] for x in X_test_missing]
X_test_missing
```

```python
X_train["Age_NA"] = X_train_missing
X_test["Age_NA"] = X_test_missing
X_train.sample(10)
```

```python
si = SimpleImputer(add_indicator=True)

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf2, y_train)

y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
